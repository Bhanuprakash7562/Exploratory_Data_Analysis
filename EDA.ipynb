```python
df_country = pd.read_excel('Country-code.xlsx')
df_country
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index #.index or .values
country_val = final_df.Country.value_counts().values
```

```python
# Pie chart - Top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.2f%%")
```

Observation : Zomato maximum transactions are from India After that USA and then UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})   
```

```python
ratings
```

## Observation
1. when the rating is 4.5-4.9 ------> Excellent
2. when the rating is 4.0-4.4 ------> Very Good
3. when the rating is 3.5-3.9 ------> Good
4. when the rating is 2.5-3.4 ------> Average
5. when the rating is 1.8-2.4 ------> Poor
6. when the rating is 0.0 ------> Not rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating color', palette=['White', 'red', 'orange', 'yellow', 'green', 'green'])
```

## Observation
1. Not rated count is very high.
2. maximum number of ratings are between 2.5-3.4

```python
## Countplot
sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
